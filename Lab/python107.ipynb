# **Problem 1: Pangram**

A pangram is a sentence using every letter of the alphabet at least once. It is case insensitive, so it doesn't matter if a letter is lower-case (e.g. k) or upper-case (e.g. K).

For this exercise, a sentence is a pangram if it contains each of the 26 letters in the English alphabet.

Example: The quick brown fox jumps over the lazy dog.

*Your task is to figure out if a sentence is a pangram.*

````python
import string

# generate list of alphabets
alphabetlist = []
alphabetlist = list(string.ascii_lowercase)

# input sentence
sentence = input("Enter a sentence: ")

# convert sentence to lower case and remove special characters
sentence = sentence.lower()
sentence = ''.join(letter for letter in sentence if letter.isalpha())

# convert sentence to a list, using set to get unqiue values
sentencelist = list(set(sentence))

# check if sentence is a panagram
if (set(sentencelist) == set(alphabetlist)):
    print("The sentence is a panagram")
else:
    print("The sentence is not a panagram")
````

# **Problem 2: Isogram**

An isogram (also known as a "non-pattern word") is a word or phrase without a repeating letter, however spaces and hyphens are allowed to appear multiple times.

Examples of isograms:

lumberjacks
background
downstream
six-year-old

The word isograms, however, is not an isogram, because the s repeats.

*Your task is to figure out if the user input is isogram*

````python
# input phrase
phrase = input("Enter a word or phrase: ")

# convert phrase to lower case and remove special characters
phrase = phrase.lower()
phrase = ''.join(letter for letter in phrase if letter.isalpha())

# initialize variables
uniquewords = set()
isIsogram = True

# check for any repeating characters
for word in phrase:
    if word in uniquewords:
        isIsogram = False
        break
    else:
        uniquewords.add(word)

# print results
if isIsogram:
    print("The word or phrase is an isogram")
else:
    print("The word or phrase is not an isogram")
````


To get around this problem, manufacturers print color-coded bands onto the resistors to denote their resistance values. Each band has a position and a numeric value.

The first 2 bands of a resistor have a simple encoding scheme: each color maps to a single number. For example, if they printed a brown band (value 1) followed by a green band (value 5), it would translate to the number 15.

In this exercise you are going to create a helpful program so that you don't have to remember the values of the bands. The program will take color names as input and output a two digit number, even if the input is more than two colors!

*From the example above:*

brown-green should return 15

brown-green-violet should return 15 too, ignoring the third color

````python
# dictionary of color with corresponding number
bandcolor = {
    "black" : "0",
    "brown" : "1",
    "red" : "2",
    "orange" : "3",
    "yellow" : "4",
    "green" : "5",
    "blue" : "6",
    "violet" : "7",
    "grey" : "8",
    "white" : "9"
}

# input color and clean it to only include colors in lowercase in a list
n = list(input("Enter colors, seperated by - : ").lower().split("-"))

# replace the color with the corresponding numbers
for i in range(len(n)):
  if n[i] in bandcolor:
    n[i] = bandcolor[n[i]]

# only print the first two colors
print(n[0]+n[1])
````

# **Problem 5: Credit Card Number**

*Your task is to Validate Credit Card Number*

Given a number determine whether or not it is valid per the Luhn formula.

The Luhn algorithm is a simple checksum formula used to validate a variety of identification numbers, such as credit card numbers and Canadian Social Insurance Numbers.

The task is to check if a given string is valid

Valid credit card number

4539 3195 0343 6467

The first step of the Luhn algorithm is to double every second digit, starting from the right. We will be doubling

4_3_ 3_9_ 0_4_ 6_6_

If doubling the number results in a number greater than 9 then subtract 9 from the product. The results of our doubling:

8569 6195 0383 3437

Then sum all of the digits:

8+5+6+9+6+1+9+5+0+3+8+3+3+4+3+7 = 80

If the sum is evenly divisible by 10, then the number is valid. This number is valid!

````python
def LuhnAlgorithm(cardnumber):
  cardnumber = cardnumber.replace(" ","")
  total=0
  isSecond = False
  for i in range((len(cardnumber)) -1,-1,-1):
    d = int(cardnumber[i])
    if (isSecond == True):
      d *= 2
    total += d // 10
    total += d % 10
    isSecond = not isSecond
  if (total % 10 == 0):
    return True
  else:
    return False

cardnumber = input("Enter credit card number: ")

if LuhnAlgorithm(cardnumber):
  print("The credit card number is valid")
else:
  print("The credit card number is not valid")
````

# **Problem 7: Employee class**

Write a Python class Employee with properties id, name, salary, and department and methods like \__init__ calculateSalary, assignDepartment and \__str__.

Sample Employee Data:

```
"E7876", "ADAMS", 50000, "ACCOUNTING"
"E7499", "JONES", 45000, "RESEARCH"
"E7900", "MARTIN", 50000, "SALES"
"E7698", "SMITH", 55000, "OPERATIONS"
```
Use 'assignDepartment' method to change the department of an employee.

Use '\__str__' method to print the details of an employee.

Use 'calculateSalary' method takes two arguments: salary and hours_worked, which is the number of hours worked by the employee. If the number of hours worked is more than 50, the method computes overtime and adds it to the salary.

Overtime is calculated as following formula:
overtime = hours_worked - 50
Overtime amount = (overtime * (salary / 50))

````python
class Employee:

    def __init__(self, id, name, salary, department):
        self.id = id
        self.name = name
        self.salary = salary
        self.department = department

    def calculateSalary(self, salary, hours_worked):
        if (hours_worked > 50):
            overtime = hours_worked - 50
            overtime_amount = overtime * (self.salary / 50)
            total = overtime_amount + self.salary
        else:
            total = salary
        return total

    def assignDepartment(self, new_department):
        self.department = new_department

    def __str__(self):
        return f"Employee Details\nID: {self.id}\nName: {self.name}\nSalary: {self.salary}\nDepartment: {self.department}"

employee1 = Employee("E7876", "ADAMS", 50000, "ACCOUNTING")
employee2 = Employee("E7499", "JONES", 45000, "RESEARCH")
employee3 = Employee("E7900", "MARTIN", 50000, "SALES")
employee4 = Employee("E7698", "SMITH", 55000, "OPERATIONS")

# calculate and print salary of employees
print(f"Salary for employee with ID Number {employee1.id} is {employee1.calculateSalary(50000, 55):.2f}")
print(f"Salary for employee with ID Number {employee2.id} is {employee2.calculateSalary(45000, 50):.2f}")
print(f"Salary for employee with ID Number {employee3.id} is {employee3.calculateSalary(50000, 44):.2f}")
print(f"Salary for employee with ID Number {employee4.id} is {employee4.calculateSalary(55000, 62):.2f}")

print("-" * 50)
# print details of employees
print(employee1)
print("-" * 50)
print(employee2)
print("-" * 50)
print(employee3)
print("-" * 50)
print(employee4)

print("-" * 50)
# change department of an employee
print(f"Before changing department:\n{employee1}")
employee1.assignDepartment("MARKETING")
print(f"After changing department:\n{employee1}")
````

# **Problem 8: Inventory class**

Write a Python class Inventory with attributes like id, productName, availableQuantity and price. Add methods like addItem, updateItem, and checkItem_details.

Use a dictionary to store the item details, where the key is the id and the value is a dictionary containing the productName, availableQuantity and price.

Sample Data:


```
{
  "97410": {
    "name": "Television",
    "availableQuantity": 20,
    "price": 1455.99
  },
  "97411": {
    "name": "Radio",
    "availableQuantity": 32,
    "price": 654.25
  }
}
```




````python
class Inventory:

  def __init__(self):
    self.items = {}

  def addItem(self, id, productName, availableQuantity, price):
    if id in self.items:
      print(f"Item with ID {id} already exists")
    else:
      self.items[id] = {
          "name" : productName,
          "availableQuantity" : availableQuantity,
          "price" : price
      }
      print(f"Item with ID {id} added successfully")

  def updateItem(self, id, productName=None, availableQuantity=None, price=None):
    if id not in self.items:
      print(f"Item with ID {id} does not exist")
    else:
      if productName is not None:
        self.items[id]["name"] = productName
      if availableQuantity is not None:
        self.items[id]["availableQuantity"] = availableQuantity
      if price is not None:
        self.items[id]["price"] = price
      print(f"Item with ID {id} updated successfully")

  def checkItem_details(self, id):
    if id not in self.items:
      print(f"Item with ID {id} does not exist")
    else:
      item = self.items[id]
      print(f'ID: {id}\nName: {item["name"]}\nAvailable Quantity: {item["availableQuantity"]}\nPrice: {item["price"]}')

inventory = Inventory()

# adding items
inventory.addItem("97410", "Television", 20, 1455.99)
inventory.addItem("97411", "Radio", 32, 624.25)

print("\nItem Details:")
# check item details
inventory.checkItem_details("97410")
print("-" * 50)
inventory.checkItem_details("97411")

print("\n")
# update items
inventory.updateItem("97410", price=1444.41)
inventory.updateItem("97411", availableQuantity=30)

print("\nItem Details after updating:")
# check updated item details
inventory.checkItem_details("97410")
print("-" * 50)
inventory.checkItem_details("97411")
````

# **Problem 9: BankAccount class**

Write a  Python class BankAccount with attributes like accountNumber, openingBalance, currentBalance dateOfOpening and customerName. Add methods like deposit, withdraw, and checkBalance.

````python
class BankAccount:

    def __init__(self, accountNumber, openingBalance, currentBalance, dateOfOpening, customerName):
        self.accountNumber = accountNumber
        self.openingBalance = openingBalance
        self.currentBalance = currentBalance
        self.dateOfOpening = dateOfOpening
        self.customerName = customerName

    def deposit(self, amount):
        self.currentBalance += amount
        print(f"Current balance is RM {self.currentBalance:.2f}")

    def withdraw(self, amount):
        if (self.currentBalance > amount):
            self.currentBalance -= amount
            print(f"Current balance is RM {self.currentBalance:.2f}")
        else:
            print("Insufficient balance")

    def checkBalance(self):
        return self.currentBalance

    def __str__(self):
        return f"""Account Details\n
Account Number: {self.accountNumber}\n
Name: {self.customerName}\n
Date of Opening: {self.dateOfOpening}\n
Opening Balance: {self.openingBalance}\n
Current Balance: {self.currentBalance}
"""

account1 = BankAccount("12341234", 500.00, 5055.50, "2022-05-01", "Peter Parker")

# print account details
print(account1)

# deposit money
account1.deposit(222.66)

# withdraw money
account1.withdraw(3377.00)

# check current balance
print(f"{account1.currentBalance:.2f}")
````

# **Problem 10: String of Parentheses**

Write a Python class to check the validity of a string of parentheses,

```
'(', ')', '{', '}', '[' and '].
```

These brackets must be closed in the correct order,
for example

```
"()" and "()[]{}" are valid
"[)", "({[)]" and "{{{" are invalid.
```

````python
class CheckParentheses:

  def __init__(self, s):
    self.s = s

  def is_valid(self):
    matchingp = {
        ")" : "(",
        "}" : "{",
        "]" : "["
    }
    matches = []
    for i in self.s:
      if i in matchingp.values():
        matches.append(i)
      elif i in matchingp.keys():
        if (matches == []) or (matchingp[i] != matches.pop()):
          return "is invalid"
      else:
        continue
    return "is valid"

inputstrings = list(input("Enter a string of parentheses, seperated by comma: ").split(","))

for inputstring in inputstrings:
  validatestring = CheckParentheses(inputstring)
  print(f"{inputstring} {validatestring.is_valid()}")
````
