# **Problem 2: Twin Primes**

Write a simple python function that returns twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes.

Pairs of primes that differ by 2. For example, 3 and 5, 5 and 7, 11 and 13, and 17 and 19 are twin primes.

```python
def FindTwinPrimes(maxnum):
  primenum = []
  for num in range(2, maxnum):
    for divisor in range(2, num):
      if (num % divisor == 0):
        break
    else:
      primenum.append(num)
  twinprimes = []
  for i in range(len(primenum) - 1):
    if (primenum[i+1] - primenum[i] == 2):
      twinprimes.append((primenum[i], primenum[i+1]))
  return twinprimes

twinprimepairs = FindTwinPrimes(1000)
for pair in twinprimepairs:
  print(f"{pair[0]} and {pair[1]}")
```

# **Problem 3: Prime Factors**

Write a simple python function that takes a number as parameter and returns  the prime factors of that number. Prime Factorization is finding which prime numbers multiply together to make the original number.

Example: prime factors of 56 - 2, 2, 2, 7

```python
def PrimeFactors(num):
  while num % 2 == 0:
    print(2, end=", ")
    num = num // 2
  for divisor in range(3, num):
    while num % divisor == 0:
      print(divisor, end=", ")
      num = num // divisor
  print(num)

num = int(input("Enter a number: "))
print("Prime Factors of", num, "-", end=" ")
PrimeFactors(num)
```

# **Problem 4: Product of Digits**

Write a function that inputs a number and returns the product of digits of that number.

```python
def ProductOfDigits(num):
  product = 1
  while (num > 0):
    product = product * (num % 10)
    num = num // 10
  return product

num = int(input("Enter a number: "))
print(f"The product of {num} is {ProductOfDigits(num)}")
```

# **Problem 5: Proper Divisors**

Write a function that takes a number as parameter. The function finds the proper divisors of that number and then finds the sum of proper divisors. Proper divisors of a number are those numbers by which the number is divisible, except the number itself.

For example proper divisors of 36 are 1, 2, 3, 4, 6, 9, 18

```python
def SumOfProperDivisors(num):
  properdivisors = []
  for divisor in range(1, num):
    if (num % divisor == 0):
      properdivisors.append(divisor)
  print(f"The proper divisors of {num} are: {properdivisors}")
  sum = 0
  for d in properdivisors:
    sum += d
  return sum

n = int(input("Enter a number: "))
sum = SumOfProperDivisors(n)
print(f"The sum of proper divisors of {n} is {sum}")
```

# **Problem 6: Perfect Numbers**

A number is called perfect if the sum of proper divisors of that number is equal to the number. For example 28 is perfect number, since 1+2+4+7+14=28. Write a program to print all the perfect numbers in a given range

```python
def PerfectNumber(minnum, maxnum):
  perfectnumlist = []
  for p in range(2, maxnum + 1):
    for divisor in range(2, p):
      if (p % divisor == 0): break
    else:
      q = (2 ** p) - 1
      for divisor in range(2, q):
        if (q % divisor == 0): break
      else:
        perfectnum = q * (2 ** (p-1))
        if ((perfectnum >= minnum) and (perfectnum <= maxnum)):
          perfectnumlist.append(perfectnum)
  return perfectnumlist

minnum = int(input("Enter the minimum value of the range: "))
maxnum = int(input("Enter the maximum value of the range: "))
pnum = PerfectNumber(minnum, maxnum)
for num in pnum:
  print(num)
```

# **Problem 7: Amicable Numbers**

Write a python function that takes 2 parameters lower and upper (range). Let the function returns pairs of amicable numbers in that range.

Two different numbers are called amicable numbers if the sum of the proper divisors of each is equal to the other number. For example 220 and 284 are amicable numbers.

For example if we call that function: amicableNumbers(1, 1000)

The function must return: **[220, 284]**


*Why they are amicable numbers ?*

Sum of proper divisors of **220** = 1+2+4+5+10+11+20+22+44+55+110 = **284**

Sum of proper divisors of **284** = 1+2+4+71+142 = **220**





```python
def SumOfProperDivisors(num):
  sum = 0
  for divisor in range(1, num):
    if (num % divisor == 0):
      sum += divisor
  return sum

def AmicableNumbers(lower, upper):
  amicablelist = []
  for n1 in range(lower, upper + 1):
    n2 = SumOfProperDivisors(n1)
    if ((n2 > n1) and (n2 <= upper)):
      if SumOfProperDivisors(n2) == n1:
        amicablelist.append([n1, n2])
  return amicablelist

lowerrange = int(input("Enter the lower range: "))
upperrange = int(input("Enter the upper range: "))
print(AmicableNumbers(lowerrange, upperrange))
```

# **Problem 8: Maximum and Minimum Numbers**

Write a python function that takes variable length parameters and returns maximum and minimum number in the parameter numbers.

For example if we call the function: maximumMinimum(10, 20, 30, 40, 50)

The function must return: **[10, 50]**

```python
def MaximumMinimum(*args):
  minnum = args[0]
  maxnum = args[0]
  for arg in args[1:]:
    if arg < minnum:
      minnum = arg
    if arg > maxnum:
      maxnum = arg
  return [minnum, maxnum]

n_input = input("Enter numbers, seperated by commas: ")
n = [int(num) for num in n_input.split(",")]
print(MaximumMinimum(*n))
```

# **Problem 9: Pascal's Triangle**

Write a simple Python function that takes a number(n) as a parameter. Then the function prints out the first n rows of Pascal's triangle. Note : Pascal's triangle is an arithmetic and geometric figure first imagined by Blaise Pascal.



pascaltriangle.png

```python
# method 1: using binomial coefficient (C(line, i) = C(line, i-1) * (line-i+1) / i)

def PascalTriangle(n):
  for i in range(1, n+1):
    for j in range(0, n-i+1):
      print(" ", end=" ")
    c = 1
    for j in range(1, i+1):
      print(" ", c, sep=" ", end=" ")
      c = c * (i - j) // j
    print()

n = int(input("Enter n rows of Pascal's Triangle: "))
PascalTriangle(n)
```

```python
# method 2: using nCr formula (nCr = n!(n-r)!r!)

from math import factorial

def PascalTriangle(n):
  for i in range(n):
    for j in range(n-i+1):
      print(end=" ")
    for j in range(i+1):
      print(factorial(i)//(factorial(j)*factorial(i-j)),end=" ")
    print()

n = int(input("Enter n rows of Pascal's Triangle: "))
PascalTriangle(n)
```

# **Problem 10: Sort Sequence of Words**

Write a simple python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

Sample Items : green-red-yellow-black-white

Expected Result : black-green-red-white-yellow

```python
def HyphenSeperatedSeq(n):
  sortedn = sorted(n)
  result = "-".join(sortedn)
  print(result)

n = input("Enter sequence of words, seperated by hyphen:")
splitn = n.split("-")
HyphenSeperatedSeq(splitn)
```
