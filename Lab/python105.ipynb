# **Problem 1: FizzBuzz**

Write a program that prints the numbers from 1 to 100. But for multiples of three, print "Fizz" instead of the number, and for the multiples of five, print "Buzz". For numbers which are multiples of both three and five, print "FizzBuzz".

```python
def PrintFizzBuzz(num):
  if ((num % 3 == 0) and (num % 5 == 0)):
    return "FizzBuzz"
  elif (num % 3 == 0):
    return "Fizz"
  elif (num % 5 == 0):
    return "Buzz"
  else:
    return num

for num in range(1, 100+1):
  print(PrintFizzBuzz(num))
```

# **Problem 2: Collatz Sequence**

Write a program that takes an integer input from the user and generates the Collatz sequence for that number. The Collatz sequence is defined as follows:

Start with a number n. If n is even, the next number is n/2. If n is odd, the next number is 3n + 1. Repeat the process until n becomes 1.

```python
n = int(input("Enter a positive integer: "))

def CollatzSeq(n):
    if (n % 2 == 0):
        n = n / 2
    else:
        n = 3 * n + 1
    print(f"{n:.0f}", end=" ")
    return n

print(f"The Collatz Sequence for {n}:")
if (n == 1):
    print(n)
elif (n > 1):
    print(n, end=" ")
    while (n != 1):
        n = CollatzSeq(n)
else:
    print("Error! Enter a positive integer.")
```

# **Problem 3: GCD Calculation**

Write a program that takes two integers from the user and calculates their greatest common divisor (GCD) using the Euclidean algorithm.

You can refer this Link

https://www.khanacademy.org/computing/computer-science/cryptography/modarithmetic/a/the-euclidean-algorithm

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

def GCD(a,b):
    while (b != 0):
        r = a % b
        a = b
        b = r
    return a

print(f"GCD({a},{b}):", end=" ")

if (a > b):
    print(GCD(a,b))
elif (b > a):
    print(GCD(b, a))
```

# **Problem 4: Rock, Paper, Scissors**

Write a program that plays the game of Rock, Paper, Scissors with the user. The user makes a choice, the program randomly chooses, and the winner is determined.

To generate random number use random module

> import random

> random.randint(1,3)


```python
import random

# input in integer
print("Let's play Rock, Paper, Scissors!")
print("Enter 1 for Rock | Enter 2 for Paper | Enter 3 for Scissors")
n = int(input("Enter 1 | 2 | 3: "))

# generate random number
r = random.randint(1,3)

# function to assign string choice to integer
def RockPaperScissors(x):
    if (x == 1):
        x = "Rock"
    elif (x == 2):
        x = "Paper"
    elif (x == 3):
        x = "Scissors"
    return x

# condition to determine winner
if ((n > 0) and (n < 4)):
    n = RockPaperScissors(n)
    r = RockPaperScissors(r)
    print(f"You chose {n}. Computer chose {r}.")
    if (n == r):
        print("Tie! Try again.")
    elif ((n == "Rock") and (r == "Scissors") or (n == "Paper") and (r == "Rock") or (n == "Scissors") and (r == "Paper")):
        print("You win!")
    else:
        print("You lose!")
else:
    print("Error! Please enter 1, 2, or 3.")
```

# **Problem 5: Number Guessing Game**

Write a program that randomly generates a number between 1 and 100. The user has to guess the number. After each guess, the program tells the user whether the guess is too high, too low, or correct. The game continues until the user guesses the correct number.

To generate random number use random module

> import random

> random.randint(1,3)

```python
import random

r = random.randint(1, 100)

n = int(input("Enter a number between 1 and 100: "))

def guess(n, r):
  if (n > r):
    return "Too high"
  elif (n < r):
    return "Too low"

while (n != r):
  print(guess(n, r))
  n = int(input("Try again: "))
print("Correct!")
```

# **Problem 6: Perfect Number**

Write a program that generates perfect numbers.

Example

6: The divisors of 6 are 1, 2, 3, and 6. The sum of its proper divisors (excluding 6 itself) is 1 + 2 + 3 = 6.

28: The divisors of 28 are 1, 2, 4, 7, 14, and 28. The sum of its proper divisors (excluding 28 itself) is 1 + 2 + 4 + 7 + 14 = 28.


```python
# q = (2 ** p) - 1, where p and q are prime numbers
# perfect number = q * (2 ** (p - 1))
# limit at 4 numbers becuase the numbers are too large

count = 4
i = 1

for p in range(2, 50):
  for divisor in range(2, p):
    if (p % divisor == 0): break
  else:
    q = (2 ** p) - 1
    for divisor in range(2, q):
      if (q % divisor == 0): break
    else:
      if (count > 0):
        perfectnum = q * (2 ** (p - 1))
        print(i, "-", perfectnum)
        count -= 1
        i += 1
```

# **Problem 7: Harmonic Series**

Write a program that calculates the sum of the first n terms of the harmonic series. Take the n as Input.

Hn = 1 +  1/2 + 1/3 + 1/4 .... + 1/n

```python
n = int(input("Enter n terms of the harmonic series: "))

def HarmonicSeries(n):
  sum = 0
  for i in range(1, n+1):
    sum = sum + 1/i
  return sum

print(f"The sum is {HarmonicSeries(n):.4f}")
```

# **Problem 8: Number to Words**

Write a program that converts a number to its word representation (e.g., 123 to "one hundred twenty-three").

```python
n = int(input("Enter a number between 0 and 1000: "))

twenty = ["", "One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten",
          "Eleven", "Twelve", "Thirteen", "Fourteen", "Fifteen", "Sixteen", "Seventeen", "Eighteen", "Nineteen"]
tens = ["", "", "Twenty", "Thirty", "Forty", "Fifty", "Sixty", "Seventy", "Eighty", "Ninety"]

if (n == 0):
  print("Zero")
elif (n < 20):
  print(twenty[n])
elif (n < 100):
  print((tens[n // 10]) + (" " + twenty[n % 10] if n % 10 != 0 else ""))
elif (n < 1000):
  print((twenty[n // 100]) + " Hundred", end =" ")
  if (n % 100 != 0):
    r = n % 100
    if (r < 20):
      print(twenty[r])
    else:
      print((tens[r // 10]) + (" " + twenty[r % 10] if r % 10 != 0 else ""))
elif (n == 1000):
  print("One Thousand")
else:
  print("Error! Enter a number between 0 and 1000.")
```

# **Problem 9: Roman to Integer**

Write a program to convert a Roman numeral to an integer and also convert integer to Roman numeral

```python
# convert roman numeral to integer

def integertoroman(roman):
  roman_symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
  num_values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
  index = 0
  num = 0
  i = 0
  while i < len(roman):
    if ((i + 1 < len(roman)) and (roman[i:i+2] in roman_symbols)):
      index = roman_symbols.index(roman[i:i+2])
      num += num_values[index]
      i += 2
    else:
      index = roman_symbols.index(roman[i])
      num += num_values[index]
      i += 1
  return num

# convert integer to roman numeral

def romantointeger(num):
  if ((num > 3999) or (num < 1)):
    print("Error! Enter a number between 0 and 3999.")
    return
  else:
    roman_symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
    num_values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    i = 0
    roman = " "
    while num > 0:
      r = num // num_values[i]
      num = num % num_values[i]
      while r > 0:
        roman = roman + roman_symbols[i]
        r -= 1
      i += 1
  return roman

# input to choose romantointeger or integertoroman

print("Enter 1 to convert a Roman numeral to an integer")
print("Enter 2 to convert an integer to a Roman numeral")
choice = int(input(": "))

if (choice == 1):
  roman = input("Enter a Roman numeral: ")
  print(f"The integer for {roman} is {integertoroman(roman)}")
elif (choice == 2):
  num = int(input("Enter a number between 0 and 3999: "))
  print(f"The Roman numeral for {num} is {romantointeger(num)}")
else:
  print("Error! Enter a valid number.")
```

# **Problem 10: String Compression**

Write a program to perform basic string compression using the counts of repeated characters (e.g., "aabcccccaaa" -> "a2b1c5a3").

```python
s = input("Enter a string: ")

c = []
count = 1

for i in range(1, (len(s))):
  if s[i] == s[i - 1]:
    count += 1
  else:
    c.append(s[i-1] + str(count))
    count = 1
c.append(s[-1] + str(count))

c_string = ""
for part in c:
  c_string += part
print(c_string)
```
