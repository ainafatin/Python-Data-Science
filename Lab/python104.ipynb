# **Problem 1: Natural Numbers**

Print first 10 natural numbers using while loop

```python
# natural numbers are whole, non-negative numbers
# initialize the first natural number (nn)
# initialize the number of integers that will print out (count)
nn = 1
count = 10

# using while loop to print results
while (count > 0):
  print(nn, end = " ")
  nn += 1
  count -= 1
```

# **Problem 2: Prime Numbers**

Print first 10 prime numbers using for loop

```python
# prime numbers are natural numbers that can be divided by 1 and itself only
# initialize the count, initial number, and maximum number
count = 10
num = 2
maxnum = 50

# using for loop to print results
for n in range(num, maxnum):
  for divisor in range(2, n):
    if (n % divisor == 0):
      break
  else:
    if (count > 0):
      print(n, end=" ")
      count -= 1
```

# **Problem 3: Adam Numbers**

Get number of items as input and generate that many Adam Numbers

```python
# prompt for input
n = int(input("Enter the number of Adam Numbers to generate: "))

# initialize the start number to check for Adam Numbers (num)
# initialize the number of iterations (count)
num = 1
count = 0

# use while loop to check the conditions and print results
while (count < n):
  # calculate square of the number
  sqnum = num * num
  # reverse the number
  revnum = 0
  tempnum = num
  while (tempnum > 0):
    revnum = revnum * 10 + tempnum % 10
    tempnum = tempnum // 10
  # calculate square of the reversed number
  revsq = revnum * revnum
  # reverse the square of the reversed number
  revsqrev = 0
  tempsq = revsq
  while (tempsq > 0):
    revsqrev = revsqrev * 10 + tempsq % 10
    tempsq = tempsq // 10
  # check if the reversed square of the reversed number is same as the square of the number
  if (sqnum == revsqrev):
    print(num, end=" ")
    count += 1
  num += 1
```

# **Problem 4: Armstrong Numbers**

Get number of items as input and generate that many Armstrong Numbers

```python
# prompt for input
n = int(input("Enter the number of Armstrong Numbers to generate: "))

# initialize the start number to check for Armstrong Numbers (num)
# initialize the number of iterations (count)
num = 1
count = 0

# use while loop to check the conditions and print results
while (count < n):
  # convert int to string to count the numbers
  strnum = str(num)
  lennum = len(strnum)
  # calculate sum of numbers to the power of the length of numbers
  sum = 0
  for sn in strnum:
    sum = sum + int(sn) ** lennum
  # check if the sum is the same as the number
  if (sum == num):
    print(num, end=" ")
    count += 1
  num += 1
```

# **Problem 8: Command Line Argument**

Write a python program that takes few numbers as command line argument. Use a loop to display all elements. Use another loop to display all elements in the even position. Use another loop to display all elements in the odd position.

```python
import sys
args = sys.argv[1:]

# display all elements
print("All elements:")
for arg in args:
    print(arg)

# display elements in the even position
print("Elements in the even position:")
for i in range(1, len(args), 2):
    print(args[i])

# display elements in the odd position
print("Elements in the odd position:")
for i in range(0, len(args), 2):
    print(args[i])
```

# **Problem 9: Command Line Argument**

Write a python program that takes few numbers as command line argument. Find the average of those numbers.

```python
import sys
args = sys.argv[1:]

# addition of the numbers
sum = 0
for i in args:
    sum = sum + int(i)

# average of the numbers
average = sum / int(len(args))

# print results
print(f"Average is {average:.2f}")
```

# **Problem 10: Are the Numbers Different?**

Write a Python program that takes a string as input, which contains numbers separated by commas. Convert the string to a list of numbers and determine whether all the numbers are different from each other.

```python
# prompt user for input
numbers = input("Enter numbers (seperate by commas): ")

# split the numbers
nums = numbers.split(",")

# convert list of strings to integer
numlist = [int(num) for num in nums]

# set the condition to be true
isDiff = True

# use nested loop to check for duplicates
for i in range(len(numlist)):
  for j in range(i+1, len(numlist)):
    if (numlist[i] == numlist[j]):
      isDiff = False
    j += 1
  i += 1

# print the results
if isDiff:
  print("All the numbers are different from each other.")
else:
  print("Some numbers are the same with each other.")
```

# **Problem 11: Sum of 100**

Write a Python program that takes a string as input, which contains numbers separated by commas. Convert the string to a list of numbers. Now pick 3 unique numbers from the list whose sum is 100.

```python
# prompt user for input
numbers = input("Enter numbers (seperate by commas): ")

# split the numbers
nums = numbers.split(",")

# convert list of strings to integer
numlist = [int(num) for num in nums]

# set the condition to be false
isUnique = False

# for loop to iterate over three numbers
for x in range(len(numlist)):
  for y in range(x+1, len(numlist)):
    for z in range(y+1, len(numlist)):
      # check if the three numbers sums up to 100
      if (numlist[x] + numlist[y] + numlist[z] == 100):
        # check if the three numbers are unique
        if ((numlist[x] != numlist[y]) and (numlist[x] != numlist[z]) and (numlist[y] != numlist[z])):
          print(f"The three unique numbers whose sum is 100 are {numlist[x]}, {numlist[y]}, and {numlist[z]}")
          isUnique = True

# if condition remains to be false
if not isUnique:
  print("No combination of unique numbers add up to 100")
```

# **Problem 12: Multiplication of Two Numbers**

# **Problem 13: Fibonacci Series**

Write a python program to print first 10 terms in a Fibonacci series

```python
# Fibonacci series is a sequence of numbers found by adding up the two numbers before it
# initialize the first and second terms (x, y)
# initialize the maximum terms to print (count)
x = 0
y = 1
count = 10

# print the first and second terms
print("The first 10 terms in a Fibonacci series are:")
print(x, end=" ")
print(y, end=" ")

# calculate and print the remaining terms
for i in range(2, count):
  z = x + y
  print(z, end=" ")
  x = y
  y = z
```

# **Problem 14: Convert to Binary**

Write a python program which takes a number as input and convert the number to binary. Note: Don't use any builtin functions.

```python
# prompt user for input
n = int(input("Enter a number: "))

# initialize value for binary conversion (empty string)
# initialize num to perform conversion operation
b = " "
num = n

# convert to binary using repeated division by 2
if (n == 0):
  b = "0"
else:
  while (num > 0):
    remainder = num % 2
    b = str(remainder) + b
    num //= 2

# print the binary conversion
print(f"The binary of {n} is {b}")
```

# **Problem 15: Convert to Decimal**

Write a python program which takes a binary number as input and convert the number to decimal. Note: Don't use any builtin functions.

```python
# prompt user for input
b = input("Enter a binary number: ")

# initialize value for decimal conversion
# initialize position number in binary
n = 0
p = len(b) - 1

# convert binary to decimal
for i in b:
  if i == '1':
    n = n + 2 ** p
  p = p - 1

# print the decimal conversion
print(f"The decimal of {b} is {n}")
```
