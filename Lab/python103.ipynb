
**Problem 1:**

A laboratory technician always needs to prepare various solutions.

Coming Sunday, he has to create a 20-liter solution that is 35% salt by mixing two available solutions. One solution (A) is 25% salt, and the other solution (B) is 50% salt. How many liters of each solution are required to achieve the desired concentration?

Coming Monday, he has to create an 8-liter solution that is 25% sugar by mixing two available solutions. One solution (A) is 15% sugar, and the other solution (B) is 40% sugar. How many liters of each solution are required to achieve the desired concentration?

Write a simple, generic Python program to assist the laboratory technician. The program must take all these numbers (20 liters, 35, 25, 50) as input and calculate the required liters of each solution and print them. Please note that the same program must work for the second problem as well (8 liters, 25, 15, 40).

The maximum stock for solution (A) and solution (B) is always 3 liter only. After calculating/printing the required quantity of A and B, throw proper message. For example, required quantity of solution (A) is less than 3 liter say "Solution (A) is available can proceed". If required quantity of solution (B) is greater than 3 liter say "Solution (B) is not available, please order 1.3 liter now".

```python
# prompt laboratory technician for input
totalsol = int(input("Enter the total amount of solution in liters: "))
totalconc = int(input("Enter the percentage of the concentration: "))
conc_a = int(input("Enter the concentration percentage of Solution A: "))
conc_b = int(input("Enter the concentration percentage of Solution B: "))

# convert percentages to decimal
conc_a = conc_a / 100
conc_b = conc_b / 100

# calculation of total concentration over total solution
amount = (totalconc / 100) * totalsol

# derive calculations for solution a and solution b based on the known formulas
# totalsol = sol_a + sol_b
# totalconc = (conc_a * sol_a) + (conc_b * sol_b)

# calculate the amount for each solution
sol_a = (amount - conc_b * totalsol) / (conc_a - conc_b)
sol_b = totalsol - sol_a

# check quantity for solution a and display results
if (sol_a <= 3):
  print("Solution A is available can proceed")
else:
  qty = sol_a - 3
  print(f"Solution A is not available, please order {qty:.2f} liter now")

# check quantity for solution b and display results
if (sol_b <= 3):
  print("Solution B is available can proceed")
else:
  qty = sol_b - 3
  print(f"Solution B is not available, please order {qty:.2f} liter now")
```

**Problem 2:**

Initialize two variables, x = 0b10101100 and y = 0b11011001.

Write Python code to:
1. Extract the lower 4 bits from x.
2. Check if y is even or odd.
3. Clear the upper 4 bits of x.
4. Check if the 5th bit of y is set.

```python
# initialize two variables
x = 0b10101100
y = 0b11011001
```

```python
# extract the lower 4 bits from x

# create a mask
mask = 0b00001111

# use the & operator to extract the lower 4 bits
# x         10101100   &
# mask      00001111
# result    00001100

# display results
print(bin(x & mask)) #0b1100
```

```python
# check if y is even or odd

if (y & 1 == 1):
  print("It is an odd number")
else:
  print("It is an even number")
```

```python
# clear the upper 4 bits of x

# create a mask
mask = 0b00001111

# use the & operator to clear the upper 4 bits (keep the lower 4 bits only)
# x         10101100   &
# mask      00001111
# result    00001100

# display results
print(bin(x & mask)) #0b1100
```

```python
# check if the 5th bit of y is set

# create a mask
mask = 0b00010000

# use the & operator to verify the 5th bit
# y         11011001   &
# mask      00010000
# result    00010000

# evaluate and display results
if (y & mask):
  print("The 5th bit of y is set")
else:
  print("The 5th bit of y is not set")
```

**Problem 3:**

A construction project requires two workers to complete. Worker A can complete the project in 12 hours, while Worker B can complete the same project in 16 hours. How long will it take for both workers to complete the project if they work together?

Another project requires Worker C, who can complete it in 8 hours, and Worker D, who can complete it in 10 hours. How long will it take for both workers to complete this project if they work together?

Write a simple, generic Python program to assist in calculating the time required for two workers to complete a project when working together. The program must take all these numbers (12, 16) as input and calculate the required time. Finally, print the result. Please note that the same program must work for the second problem as well (8, 10).

```python
# prompt user for input
time1 = int(input("How many hours does it take for the first worker to complete the project? "))
time2 = int(input("How many hours does it take for the second worker to complete the project? "))

# calculate the work rate of each worker (amount of work done / time taken)
workrate1 = 1 / time1
workrate2 = 1 / time2
totalworkrate = workrate1 + workrate2

# calculate the total combined time to complete project
time = 1 / totalworkrate

# display results
print(f"{time:.2f} hours is required for two workers to complete a project together.")
```

**Problem 4:**

Initialize two variables, a = 0b10101000 and b = 0b01010100.

Write Python code to:
1. Set the lower 4 bits of a.
2. Combine the bits of a and b using OR.
3. Create a mask to set the 2nd and 6th bits of a.

```python
# initialize two variables
a = 0b10101000
b = 0b01010100
```

```python
# set the lower 4 bits of a

# create a mask
mask = 0b00001111

# use the | operator to set the lower 4 bits
# a         10101000   |
# mask      00001111
# result    10101111

# display results
print(bin(a | mask)) # 0b10101111
```

```python
# combine bits of a and b using OR

# use the | operator to combine the bits
# a       10101000   |
# b       01010100
# result  11111100

# display results
print(bin(a | b)) # 0b11111100
```

```python
# create a mask to set the 2nd and 6th bits of a

# create a mask
mask = 0b00100010

# use | operator to set the 2nd and 6th bits
# a       10101000   |
# mask    00100010
# result  10101010

# display results
print(bin(a | mask)) # 0b10101010
```

**Problem 5:**

An investor decides to invest a total of RM30,000 in two different accounts. The first account offers a 5% annual interest rate, while the second account offers a 7% annual interest rate. If the total annual interest earned is RM1,800, how much money was invested in each account?

Another investor decides to invest a total of RM50,000 in two different accounts. The first account offers a 3% annual interest rate, while the second account offers a 6% annual interest rate. If the total annual interest earned is RM2,400, how much money was invested in each account?

Write a simple, generic Python program to assist in calculating the amount of money invested in each account to achieve the desired total annual interest. The program must take all these numbers (30000, 5, 7, 1800) as input and calculate the required amounts. Finally, print the result. Please note that the same program must work for the second problem as well (50000, 3, 6, 2400).

```python
# prompt user for input
totalamt = int(input("Enter total amount of investment (RM): "))
interest1 = int(input("Enter the annual interest rate for the first account (%): "))
interest2 = int(input("Enter the annual interest rate for the second account (%): "))
totalinterest = int(input("Enter the total annual interest earned (RM): "))

# convert percentage to decimal
interest1 = interest1 / 100
interest2 = interest2 / 100

# derive calculations for account1 and account2 based on the known formulas
# totalamt = account1 + account2
# totalinterest = (interest1 * account1) + (interest2 * account2)

# calculate the amount for each account
account1 = (totalinterest - interest2 * totalamt) / (interest1 - interest2)
account2 = totalamt - account1

# display results
print(f"The amount invested in the first account is RM{account1:.2f}")
print(f"The amount invested in the second account is RM{account2:.2f}")
```

**Problem 6:**

Initialize two variables, x = 0b10101100 and y = 0b11010010.

Write Python code to:
1. Swap the values of x and y without using a temporary variable.
2. Toggle the 3rd and 5th bits of x.
3. Check if two given numbers a = 29 and b = 15 are different.

```python
# initialize two variables
x = 0b10101100
y = 0b11010010
```

```python
# swap values of x and y

print(f"Before swapping, x is {bin(x)} and y is {bin(y)}")

# x         10101100   ^  x
# y         11010010
# result    01111110

# x         01111110   ^  y
# y         11010010
# result    10101100

# x         01111110   ^  x
# y         10101100
# result    11010010

x = x ^ y
y = x ^ y
x = x ^ y

print(f"After swapping, x is {bin(x)} and y is {bin(y)}")
```

```python
# toggle 3rd and 5th bits of x

# initialize x
x = 0b10101100

# create a mask
mask = 0b00010100

# x         10101100   ^
# mask      00010100
# result    10111000

# display results
print(bin(x^mask)) #0b10111000
```

```python
# check if two given numbers are different

a = 29
b = 15

# use ^ operator to check if the numbers are different
# if the numbers are the same, the binary will become 0

if (a ^ b != 0):
  print("The numbers are different")
else:
  print("The numbers are the same")
```
